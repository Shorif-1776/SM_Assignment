# Matplotlib and Seaborn for Data Science

## Task 1: Basic Matplotlib (3 plots)

```python
# Import library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Dataset
gym = pd.read_csv('gym_membership.csv')

# check data
print(gym.head)
```

```python
# print dataset shape
print(gym.shape)

# print column heading

print(gym.columns.tolist)

```

```python
# Print dataframe description

print('\nDataFrame description:')
print(gym.describe())
```

```python
# Basic DataFrame information
print('\nDataFrame info:')
print(gym.info())
```

```python
# Display last 5 rows

print('\nLast 5 rows of DataFrame')
print(gym.tail())

# Check data type

print('\nData Types:')
print(gym.dtypes)
```

### 1. Bar chart of gym membership types

```python
# Use: gym['abonoment_type'].value_counts()
# Count abonoment_type
abonoment_type_count = gym['abonoment_type'].value_counts()

# Plot bar chart
plt.bar(abonoment_type_count.index, abonoment_type_count.values)
plt.xlabel('Membership Type')
plt.ylabel('Count')
plt.title('Bar Chart of Gym Membership Types')
plt.show()
```

###2. Histogram of time in gym

```python
# Use: plt.hist(gym['avg_time_in_gym'])
# Plot Histogram
plt.hist(gym['avg_time_in_gym'])
plt.xlabel('Time in Gym')
plt.ylabel('Frequency')
plt.title('Histogram of Time in Gym')
plt.show()
```

### 3. Scatter plot: Age vs visits per week

```python
# Use: plt.scatter(gym['Age'], gym['visit_per_week'])
# Plot scatter plot
plt.scatter(gym['Age'], gym['visit_per_week'])
plt.xlabel('Age')
plt.ylabel('Visits per Week')
plt.title('Scatter Plot: Age vs Visits per Week')
plt.show()
```

### 1. Box plot: Gender vs Age

```python
# Use: sns.boxplot(data=gym, x='gender', y='Age')
# Plot box plot
sns.boxplot(data=gym, x='gender', y='Age')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Box Plot: Gender vs Age')
plt.show()
```

### 2. Count plot: Days per week

```python
# Use: sns.countplot(data=gym, x='days_per_week')
#Plot count plot
#Format xaxis
# x axis don't show clearly

plt.xticks(rotation=45)
sns.countplot(data=gym, x='days_per_week')
plt.xlabel('Days per Week')
plt.ylabel('Count')
plt.title('Count Plot: Days per Week')
plt.show()
```

```python
# Split the strings in 'days_per_week' and count individual days
days_list = gym['days_per_week'].str.split(', ').explode()
day_counts = days_list.value_counts()

# Plot the counts of individual days
plt.figure(figsize=(10, 6)) # Adjust figure size for better readability
plt.bar(day_counts.index, day_counts.values)
plt.xlabel('Day of the Week')
plt.ylabel('Count')
plt.title('Count of Visits per Day of the Week')
plt.xticks(rotation=45, ha='right') # Rotate labels again for clarity
plt.tight_layout() # Adjust layout to prevent labels from being cut off
plt.show()
```

### 3. Bar plot: Average visits by membership type

```python
# Use: sns.barplot(data=gym, x='abonoment_type', y='visit_per_week')
# Plot bar plot
sns.barplot(data=gym, x='abonoment_type' , y='visit_per_week')
plt.xlabel('Membership Type')
plt.ylabel('Average Visits per Week')
plt.title('Bar Plot: Average Visits by Membership Type')
plt.show()
```
